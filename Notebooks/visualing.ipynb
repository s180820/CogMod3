```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path= "../Data/Homework3pt1DataCSV.txt"

data = np.loadtxt(data_path)
data
```

```python
names = ['(0 -)', '(30 -)', '(- 0)', '(- 30)', '(0 0)', '(0 30)', '(30 0)', '(30 30)']

def seaborn_probability_histogram(data, bins=8, title="Histogram of Data", xlabel="Data", ylabel="Probability"):
    """Plot a histogram of the data with the specified number of bins"""
    sns.distplot(data, bins=bins, kde=False, norm_hist=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xlim([-30, 60])
    plt.ylim([0, 0.23])
    #plt.show()

#for i in range(len(data)):
 #   #plt.subplot(2, i%2, i+1)
  #  seaborn_probability_histogram(data[i], title="Histogram of Data for Stimulus pair {}".format(names[i]))
```

```python
def get_mean_and_std(data):
    """Return the mean and standard deviation of the data"""
    return np.mean(data), np.std(data)

means = []
stds = []
for i in data:
    mean, std = get_mean_and_std(i)
    means.append(mean)
    stds.append(std)
```

```python
#set figsize
plt.figure(figsize=(20, 10))
for i in range(len(data)):
    #plt subplots of 2 rows and 4 columns
    plt.subplot(2, 4, i+1)
    plt.axvline(x=means[i])
    seaborn_probability_histogram(data[i], title="Histogram of Data for Stimulus pair {}".format(names[i]))
```

```python
def guassian_probability_density_function(x, mu, sigma):
    """Calculate the probability density function of a guassian distribution"""
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-((x-mu)**2)/(2*sigma**2))

def plot_guassian_probability_density_function(mu, sigma, title="Guassian Probability Density Function", xlabel="Data", ylabel="Probability"):
    """Plot the guassian probability density function"""
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, guassian_probability_density_function(x, mu, sigma))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

plt.figure(figsize=(20, 10))
for i in range(len(data)):
    plt.subplot(2, 4, i+1)
    plot_guassian_probability_density_function(means[i], stds[i], title="Guassian Probability Density Function for Stimulus pair {}".format(names[i]))
    seaborn_probability_histogram(data[i], bins=10, title="Histogram of Data for Stimulus pair {}".format(names[i]))
```

```python
def enhancement_effect(data):
    """Calculate the enhancement effect of the data"""
    return (np.max(data) - np.min(data))/np.min(data)

def ventriliquist_illusion(data):
    """Calculate the ventriliquist illusion of the data"""
    return (np.max(data) - np.min(data))/np.max(data)

for i in range(len(data)):
    print("The enhancement effect of stimulus pair {} is {}".format(names[i], enhancement_effect(data[i])))
    print("The ventriliquist illusion of stimulus pair {} is {}".format(names[i], ventriliquist_illusion(data[i])))
```
