```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path= "../Data/Homework3pt1DataCSV.txt"

data = np.loadtxt(data_path)
data
```

```python
names = ['(0 -)', '(30 -)', '(- 0)', '(- 30)', '(0 0)', '(0 30)', '(30 0)', '(30 30)']

def seaborn_probability_histogram(data, bins=10, title="Histogram of Data", xlabel="Data", ylabel="Probability"):
    """Plot a histogram of the data with the specified number of bins"""
    sns.distplot(data, bins=bins, kde=False, norm_hist=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

for i in range(len(data)):
    #plt.subplot(2, i%2, i+1)
    seaborn_probability_histogram(data[i], title="Histogram of Data for Stimulus pair {}".format(names[i]))
```

```python
def get_mean_and_std(data):
    """Return the mean and standard deviation of the data"""
    return np.mean(data), np.std(data)

means = []
stds = []
for i in data:
    mean, std = get_mean_and_std(i)
    means.append(mean)
    stds.append(std)
```

```python
def plot_mean_and_std(means, stds, title="Mean and Standard Deviation of Data", xlabel="Stimulus Pair", ylabel="Data"):
    """Plot the mean and standard deviation of the data"""
    plt.errorbar(range(len(means)), means, yerr=stds, fmt='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(range(len(means)), names)
    plt.show()
```
